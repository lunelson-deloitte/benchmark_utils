```python
def workflow_original(readin_func: callable, input_path: str, fraction: float = 1.0):
    return f"{readin_func}('{input_path}').sample(fraction={fraction})"

def workflow_tidy(readin_func: callable, input_path: str, fraction: float = 1.0):
    return f"{readin_func}('{input_path}').sample(fraction={fraction})"


benchmark_workflows = bu.benchmark_workflows(
    workflows={
        'Original': 'workflow_original',
        'Tidy': 'workflow_tidy'
    },
    _globals=globals()
)


benchmark_operations = bu.benchmark_operations(
    operations=['join', 'filter', 'select'],
    workflows={
        'Original': 'workflow_original',
        'Tidy': 'workflow_tidy'
    },
    _globals=globals()
)
```
